```python
import pandas as pd

df = pd.read_csv("hospital_emergency_data.csv")
print(df.head())
```

```python
print("Mean:", df["emergency_cases"].mean())
print("Median:", df["emergency_cases"].median())
print("Variance:", df["emergency_cases"].var())
print("Standard Deviation:", df["emergency_cases"].std())
```

```python
hourly_avg = df.groupby("hour")["emergency_cases"].mean()
print(hourly_avg)
```

```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd

# Example data
hourly_avg = pd.Series([5, 8, 12, 15, 20, 18, 10, 6], index=range(8))
```

```python
plt.figure(figsize=(8,5))
plt.plot(hourly_avg.index, hourly_avg.values, marker='o')
plt.xlabel("Hour of Day")
plt.ylabel("Average Emergency Cases")
plt.title("Average Emergency Arrivals by Hour")
plt.grid(True)
plt.show()
```

```python
plt.plot(hourly_avg.index, hourly_avg.values, marker='o')
```

```python
import seaborn as sns

sns.lineplot(x=hourly_avg.index, y=hourly_avg.values, marker='o')
plt.xlabel("Hour of Day")
plt.ylabel("Average Emergency Cases")
plt.title("Average Emergency Arrivals by Hour")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(hourly_avg.index.astype(str), hourly_avg.values, marker='o')  # Convert index to string if categorical
plt.xlabel("Hour of Day")
plt.ylabel("Average Emergency Cases")
plt.title("Average Emergency Arrivals by Hour")
plt.grid(True)
plt.show()
```

```python
hour_variance = df.groupby("hour")["emergency_cases"].var()
print(hour_variance)
```
